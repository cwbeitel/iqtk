<h1 style="color:#9E9E9E">MDV method comparison</h1>

```python
import pandas as pd

def _read(path):
    """Read MDV comparison table, stack methods as factors for seaborn factor plot."""
    compounds = []
    methods = []
    measurements = []
    with open(path, 'r') as f:
        header = f.readline().strip().split('\t')[1:]
        for line in f:
            arr = line.strip().split('\t')
            for i, val in enumerate(arr[1:]):
                methods.append(header[i])
                compounds.append(''.join(arr[0].split(' ')))
                measurements.append(float(val))
    return pd.DataFrame({"compound": compounds, "method": methods, "measurement": measurements})

comp = 'inquiry/toolkit/metabolomics/test/mdv_comparison2.tab'
mdv_data = _read(comp)
mdv_data
```

```python
import seaborn as sns

%matplotlib inline
sns.set(style="whitegrid")

# Draw a nested barplot to show survival for class and sex
g = sns.factorplot(x="compound", y="measurement", hue="method", data=mdv_data,
                   size=14, aspect=2, kind="bar", palette="muted")

g.despine(left=True)

g.set_ylabels("Relative abundance")
g.set_xlabels("Isotopolog")
```
